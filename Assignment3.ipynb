##3) Scatter Plot

###a) Generate two lists of 20 random integers each (for x and y coordinates).

```python
import random
x = [random.randint(0, 100) for _ in range(20)]
y = [random.randint(0, 100) for _ in range(20)]
print("The lists are :")
print("X Coordinates:", x)
print("Y Coordinates:", y)
```

###b) Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
plt.scatter(x,y,c='red',s=20)
labels = ['A', 'B', 'C', 'D', 'E','F','G','H','I','J','K','L','M','N','O','P','Q','R','S','W']
for i in range(len(x)):
    plt.annotate(labels[i], (x[i], y[i]),textcoords="offset points", xytext=(5, 5))
plt.title("Scatter Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

##5) Pie Chart

###a) Visualize the percentage distribution of five categories [ 'A', 'B', 'C', 'D', 'E' ] with values [ 20, 30, 25, 15, 10 ].

```python
c = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.pie(values, labels=c)
plt.show()
```

###b) Display the percentage values on the chart and use a color palette for differentiation.

```python
col = plt.cm.Paired(np.linspace(0, 1, len(c)))
plt.pie(values, labels=c, autopct='%1.1f%%', colors=col)
plt.show()
```

##8) Comparison of Matplotlib and Seaborn

###a) Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn.

```python
x1 = ['A', 'B', 'C', 'D', 'E']
y1 = [40, 8, 10, 15, 25]

plt.subplot(1, 2, 1)
plt.plot(x1, y1, marker='o',)
plt.title('Line Plot (Matplotlib)')

plt.subplot(1, 2, 2)
sns.lineplot(x=x1, y=y1, marker='o', color='orange')
plt.title('Line Plot (Seaborn)')

plt.tight_layout()
plt.show()
```

###b) Highlight the differences in the code, syntax, and aesthetics of the two libraries.


<h2>Code Syntax :</h2>
<h4>Matplotlib:</h4>
 More verbose, requiring explicit calls for plot elements (title, labels, colors).

1)Uses plt.plot() for line plots.

2)Input data is provided as positional arguments (categories, values).

3)Customization requires separate commands like plt.title, plt.xlabel, and plt.ylabel  
<h4>Seaborn:</h4>
 Higher-level functions, often needing less code to create visually appealing plots.  Seaborn can automatically handle many of the details Matplotlib needs explicit calls for.  For example, color palettes, labels etc.

1) Uses sns.lineplot() for line plots.

2) Input data must be explicitly specified as keyword arguments (x=categories, y=values).

3) Many styling features, such as gridlines, are applied automatically.

##Aesthetics :
####Matplotlib:
Basic, customizable, but requires effort to create visually appealing plots.  The default looks can be quite plain.
####Seaborn:
Built on Matplotlib but provides improved aesthetics and default styles.  Seaborn automatically produces more visually appealing graphs.
