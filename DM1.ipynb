```python
import pandas as pd
data=pd.read_csv("/kaggle/Iris.csv")
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,6))
sns.boxplot(x="SepalLengthCm",y="SepalWidthCm",data=data,palette="Set2")
plt.show()
sns.barplot(x="PetalLengthCm",y="PetalWidthCm",data=data,palette="hls")
plt.show()
sns.histplot(x="SepalLengthCm",y="PetalLengthCm",data=data,color="gold")
plt.show()
sns.scatterplot(x="PetalWidthCm",y="SepalWidthCm",data=data,color="red")
plt.show()
flowers=pd.Series(data['Species']).value_counts()
sizes=flowers.values
labels=flowers.index
c=['red','green','blue']
plt.pie(sizes,labels=labels,colors=c)
plt.show()
```
